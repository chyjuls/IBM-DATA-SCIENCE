We first need to create a folium `Map` object, with an initial center location to be NASA Johnson Space Center at Houston, Texas.

We could use `folium.Circle` to add a highlighted circle area with a text label on a specific coordinate. For example,

```python
# Initial the map
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)


# For each launch site, add a Circle object based on its coordinate (Lat, Long) values. In addition, add Launch site name as a popup label


# Coordinates for each launch site
launch_sites = [
    {"name": "CCAFS LC-40", "lat": 28.562302, "long": -80.577356},
    {"name": "CCAFS SLC-40", "lat": 28.563197, "long": -80.576820},
    {"name": "KSC LC-39A", "lat": 28.573255, "long": -80.646895},
    {"name": "VAFB SLC-4E", "lat": 34.632834, "long": -120.610745}
]

# Initial the map with a central coordinate
central_lat = sum([site['lat'] for site in launch_sites]) / len(launch_sites)
central_long = sum([site['long'] for site in launch_sites]) / len(launch_sites)
site_map = folium.Map(location=[central_lat, central_long], zoom_start=5)

# For each launch site, add a Circle object and a Marker
for site in launch_sites:
    circle = folium.Circle([site['lat'], site['long']], radius=1000, color='#d35400', fill=True).add_child(folium.Popup(site['name']))
    marker = folium.Marker(
        [site['lat'], site['long']],
        # Create an icon as a text label
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % site['name'],
        )
    )
    site_map.add_child(circle)
    site_map.add_child(marker)

# Display the map
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
from folium.features import DivIcon
from folium.plugins import MarkerCluster

# Initialize site_map if not already initialized
site_map = folium.Map(location=[28.562302, -80.577356], zoom_start=5)

# Create a MarkerCluster object
marker_cluster = MarkerCluster()

# Assume spacex_df is your DataFrame and has been defined earlier in the code
# Iterate through the DataFrame and add each launch as a marker to the cluster
for index, record in spacex_df.iterrows():
    marker = folium.Marker(
        [record['Lat'], record['Long']],
        icon=folium.Icon(color='white', icon_color=record['marker_color']),
        popup=f"{record['Launch Site']}, {record['Date']}, {record['Payload']}"
    )
    marker_cluster.add_child(marker)

# Add MarkerCluster to the map
site_map.add_child(marker_cluster)

# Example coordinates for the closest coastline point (replace with actual values)
coastline_coord = [28.562302, -80.567356]  # Closest coastline point to the launch site
launch_site_coord = [28.573255, -80.646895]  # Coordinates for the launch site
distance = 0.9  # Distance from the launch site to the coastline point in kilometers

# Add a marker for the coastline point
distance_marker = folium.Marker(
    coastline_coord,
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html=f'<div style="font-size: 12; color:#d35400;"><b>{distance:.2f} KM</b></div>',
    )
)
site_map.add_child(distance_marker)

# Draw a PolyLine between the launch site and coastline point
line = folium.PolyLine(locations=[launch_site_coord, coastline_coord], weight=1, color='blue')
site_map.add_child(line)

# Optionally, add a marker for the launch site itself if not already included in spacex_df
launch_site_marker = folium.Marker(
    launch_site_coord,
    icon=folium.Icon(color='red', icon='rocket'),
    popup='Launch Site'
)
site_map.add_child(launch_site_marker)

# Display the map
site_map
```
